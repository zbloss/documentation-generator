```python
import os
import re
import numpy as np
import pandas as pd

import nbformat
from nbformat import current
from nbconvert import PythonExporter

from tqdm.notebook import tqdm
```

## Gathering My Code

Here I'm gathering a list of all of the python files and jupyter notebooks that I have under my project directory.

```python
pattern = ".py"
mypath = "C:/Users/altoz/Documents/Projects/"
py_files = []
nb_files = []
for root, dirs, files in os.walk(mypath):
    for file in files:
        if file.endswith('.py'):
            if not file.endswith('__init__.py') \
                and not file.endswith('setup.py') \
                and not file.endswith('__version__.py'):
                
                py_files.append(root.replace('\\', '/') + f'/{file}')
        elif file.endswith('.ipynb'):
            if not '.ipynb_checkpoints' in file:
                nb_files.append(root.replace('\\', '/') + f'/{file}')
```

```python
print(f'Python Files: {len(py_files)}')
print(f'Jupyter Notebooks: {len(nb_files)}')
```

## Converting Jupyter Notebooks

I do a lot of my work in jupyter notebooks before converting it into python scripts. So I'll build a process to recursively convert each notebook into a script and store it in my $D:$ drive

```python
def convert_nb_to_py(nb_filepath, new_filepath, exporter):
    """
    Given the path to a jupyter notebook, converts it to a python file
    and stores it in new_filepath.
    
    :param nb_filepath: (str) path to the jupyter notebook.
    :param new_filepath: (str) path to the new python file.
    :param exporter: The exporter you wish to use.
    :returns: None.
    """

    with open(nb_filepath, 'r') as nb_file:
        notebook = nbformat.reads(nb_file.read(), as_version=4)
        nb_file.close()
        
    source, _ = exporter.from_notebook_node(notebook)
    with open(new_filepath, 'w+') as py_file:
        py_file.writelines(source)
        py_file.close()
    
    return None
```

```python
exporter = PythonExporter()
base_d_path = 'D:/data/documentation-generator/py_files/'

for nb_file in tqdm(nb_files, desc='Converting Notebooks...'):
    nb_file = nb_file.replace('//', '/')
    py_filename = nb_file.split('/')[-1].split('.')[0] + '.py'
    py_filename = f'{base_d_path}/{py_filename}'
    
    try:
        convert_nb_to_py(nb_file, py_filename, exporter)
    except:
        pass        
```

## Copying Python Files

```python
for py_file in tqdm(py_files, desc='Copying Python Files...'):
    new_filename = base_d_path + py_file.split('/')[-1]

    try:
        with open(py_file, 'r') as in_file:
            data = in_file.read()
            in_file.close()

        with open(new_filename, 'w') as out_file:
            out_file.writelines(data)
            out_file.close()
    except:
        pass
```
